### Data Preparation
#### Promotion period on weekly basis 2016-2017 each week, calculate Adstock & reach for each vehicle for each week

```python
# TV 8 week half-life; radio 4 week half-life, calculate alpha
a_tv = 1-0.5**(1/8)
a_radio = 1-0.5**(1/4)
```

Adstock & Reach for TV

```python
tv = promo[promo['vehicle'] =='TV']
tv
```

```python
# merge tv promo with trans
week = pd.DataFrame(promo['tran_wk']).reset_index()
week = week.drop(['index'],axis=1)
tv_promo = pd.merge(week,tv,on='tran_wk', how='left')
tv_promo['adstock'] = 0
tv_promo['reach'] = 0
tv_promo = tv_promo.fillna(0)
tv_promo = tv_promo.drop_duplicates()
tv_promo
```

```python
tv_promo[tv_promo['vehicle']=='TV']
```

```python
# compute adstock and reach
# Adstock(t) = aGRP(t) + (1-a)Adstock(t-1)
# Reach for TV: Reach = 0.95(1-e^(-0.020GRP))
import numpy as np

for i in range(106):
        tv_promo.iloc[i,5] = a_tv * tv_promo.iloc[i,2] + (1-a_tv)*tv_promo.iloc[i-1,5]
        tv_promo.iloc[i,6] = 0.95 * (1-np.exp(-0.02*tv_promo.iloc[i,5]))
```

```python
tv_promo.describe()
```

```python
# rename columns for furthur data frames merge
tv_promo = tv_promo.rename(columns={
    "reach":"TV",
    "adstock":"TV_adstock"
    
})

# only select specific columns
tv_promo = tv_promo[[
    'tran_wk','TV_adstock','TV'   
]]

tv_promo.head()
```

Adstock & Reach for Radio

```python
radio = promo[promo['vehicle'] =='Radio']
radio
```

```python
radio_promo = pd.merge(week,radio,on='tran_wk', how='left')
radio_promo['adstock'] = 0
radio_promo['reach'] = 0
radio_promo = radio_promo.fillna(0)
radio_promo = radio_promo.drop_duplicates()
radio_promo.head()
```

```python
radio_promo['amount'].sum()
```

```python
# compute adstock and reach
# Adstock(t) = aGRP(t) + (1-a)Adstock(t-1)
# Reach for Radio: Reach = 0.90(1-e^(-0.020GRP))
import numpy as np

for i in range(106):
        radio_promo.iloc[i,5] = a_radio * radio_promo.iloc[i,2] + (1-a_radio)*radio_promo.iloc[i-1,5]
        radio_promo.iloc[i,6] = 0.90 * (1-np.exp(-0.02*radio_promo.iloc[i,5]))
```

```python
radio_promo.describe()
```

```python
# rename columns for furthur data frames merge
radio_promo = radio_promo.rename(columns={
    "reach":"Radio",
    "adstock":"Radio_adstock"
})

# only select specific columns
radio_promo = radio_promo[[
    'tran_wk','Radio_adstock','Radio'   
]]

radio_promo.head()
```

```python
# merge TV and Radio
tv_radio = pd.merge(tv_promo,radio_promo, on='tran_wk')
tv_radio
```

#### Weekly transaction details for each product

Label (target variable):

* Weekly sales quantity


Product Features: 

* Shelf Price

* Discount Rate


```python
# convert tran_dt to tran_wk to let the promo starts on Sunday for a week
trans['tran_wk'].value_counts()
```

```python
list(trans.columns)
```

```python
# average unit price and discount rate of products weekly
trans_weekly = trans.groupby(['tran_wk','prod_id']).agg({"tran_prod_sale_qty":"sum","prod_unit_price":"mean","discount_rate":"mean"}).reset_index()
trans_weekly = trans_weekly.rename(columns={"tran_prod_sale_qty":"weekly_sale_qty","prod_unit_price":"weekly_shelf_price",
                                                 "discount_rate":"weekly_discount_rate"})
trans_weekly['tran_wk'] = pd.to_datetime(trans_weekly['tran_wk'])
trans_weekly.sort_values(by='tran_wk',ascending=True)
```

Include Seasonality & Holiday in the Model

```python
# seasonality
season['tran_wk'] = pd.to_datetime(season['tran_wk'])
holiday['tran_wk'] = pd.to_datetime(holiday['tran_wk'])

# season
weekly_season = pd.merge(trans_weekly,season, on='tran_wk', how='left')

# holiday
weekly_holiday = pd.merge(weekly_season,holiday, on='tran_wk', how='left')
weekly_holiday = weekly_holiday.fillna(0)
weekly_holiday
```
